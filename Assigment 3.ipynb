```
from pyspark.sql.functions import *
```

```
# ----a)	Create HR_Employee Pysaprk dataFrame.


HR_Employee_DF = spark.read.csv("dbfs:/FileStore/tables/HR_Employee.csv", header = True, inferSchema = True)
# type(HR_Employee_DF)

HR_Employee_DF.show()
```

```
#-- b)	Show printSchema & show 10 rows of Dataset

HR_Employee_DF.printSchema()
HR_Employee_DF.show(10)
```

```

HR_Employee_DF.createOrReplaceTempView("employee_table")
```

```
# --c)	Most employee is working in which department

spark.sql("select Department ,count(*) as most_num_emp from employee_table group by Department order by most_num_emp  desc ").show(1)
```

```
#---- d)Highest number of job roles

spark.sql("select Max(JobRole) as Job_role,count(*) as most_num_emp from employee_table ").show()
```

```
# -----e)    Which gender have higher strength as workforce?

spark.sql("select Max(Gender) as Gender,count(*) as most_Strength from employee_table").show()
```

```
#--- f) Compare the marital status of employee and find the most frequent status.

spark.sql("select MaritalStatus as Gender,count(*) as most_freq_status from employee_table where BusinessTravel ='Travel_Frequently' group by MaritalStatus ").show()
```

```
# g) -----  Mostly hired employee have qualification


spark.sql("select Education, count(*) as Number_of_Qualified_Employee from employee_table  group by Education order by Number_of_Qualified_Employee desc").show()
```

```
# h)------Find the count of employee from which education fields


spark.sql("select EducationField, count(*) as Number_of_Employee from employee_table group by EducationField order by Number_of_Employee desc").show()
```

```
# i)----- What is the job satisfaction level of employee?



spark.sql("select    JobSatisfaction , count(*) as Number_of_Employee from employee_table  group by JobSatisfaction order by Number_of_Employee desc").show()
```

```
# j) -----   Does most of employee do overtime: Yes or No?


spark.sql("select Overtime, count(*) as Number_of_Employee from employee_table where OverTime ='Yes' or Overtime ='No' group by Overtime ").show()
```

```
# m)  ----  Find the AVG salary of graduate employee.



spark.sql("select mean(Income) as Averge_salary_employee  from employee_table where Education = 'Bachelor' ").show()
```

```
#n)-----	Find the employee qualification receiving salary lower than equal to avg. salary of all employee. 

spark.sql("select Income, Education from employee_table where Income <= (select avg(Income) from employee_table)").show()
```

```
#o)-----	When does the employee have highest chance of promotion in terms of working year


spark.sql("select Performance_Rating, count(*) as Promotion_chances from employee_table group by Performance_Rating order by Promotion_chances desc ").show()
```

```
#p)-----	Highest attrition is in which department?


spark.sql("select Department, count(*) as total_count from employee_table where Attrition='Yes' group by Department order by total_count desc ").show(1)
```

```
#q)------	Show marital status of Person having highest attrition rate

spark.sql("select MaritalStatus, count(*) as no_of_HighAttrition from employee_table where Attrition='Yes' group by MaritalStatus order by no_of_HighAttrition  desc ").show(1)
```
